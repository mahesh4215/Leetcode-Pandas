Filter Function

```python
import pandas as pd
df = pd.read_csv('/content/sample_data/world_population.csv')
print(df)
```

```python
df[df['Rank'] > 100]
```

```python
df[df['Country'].isin(['Afghanistan', 'Albania', 'Zimbabwe'])]
```

```python
df2 = df.set_index('Country')
```

```python
df2.filter(items = ['Rank','Continent'])
```

```python
df2.filter(like = 'United', axis = 0)
```

```python
df2.loc['United States']
```

Filter by NOT in list (~isin)

```python
df[df['Country'] == 'India']
```

```python
df[~df['Country'].isin(['India', 'Germany'])]
```

```python
df[df['Country'].str.startswith('A')]
df[df['Country'].str.endswith('land')]
```

```python
df[(df['Continent'] == 'Asia') & (df['2022 Population'] > 1000000)]
df[(df['Continent'] == 'Europe') | (df['Capital'] == 'Paris')]
```

Filter by Null/NotNull values

```python
df[df['Capital'].isnull()] #--- check  any null values in Capital Column
df[df['Capital'].notnull()] # check not null values
```

```python
df.query("Continent == 'Asia' and `2022 Population` > 1000000")
```

group by and aggregating

```python
import pandas as pd
df = pd.read_csv('/content/sample_data/Flavors.csv')
print(df)
```

```python
df.groupby('Base Flavor').mean(numeric_only = True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','max']})
```

Merge, Join and concat

```python
import pandas as pd
df1 = pd.read_csv('/content/sample_data/LOTR.csv')
df1
```

```python
df2 = pd.read_csv('/content/sample_data/LOTR 2.csv')
df2
```

```python
print(df1.merge(df2, how='inner', on='FellowshipID').to_string(index=False))
```

```python
df1.merge(df2, how = 'left')
```

```python
pd.concat([df1,df2])
```
